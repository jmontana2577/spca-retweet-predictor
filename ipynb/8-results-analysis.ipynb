#### Establish Baseline Accuracy
I'm going to see which actual retweet number appears most in the samples and calculate the percentage of samples with that number. This will serve as the baseline accuracy (meaning the accuracy if I just guess the number of retweets that appears most).

```python
df.groupby(by='Actual', as_index=False).count()[['Actual','Predicted']].\
rename(columns={'Actual':'Retweets','Predicted':'Count'}).\
sort_values('Count', ascending=False).head(2)
```

```python
# calculate the percentage of tweets with 0 retweets
(df.Actual == 0).sum() / df.shape[0]
```

**Baseline:**  
If I were to guess 0 retweets for each tweet in the test dataset, I would be correct 24.9% of the time.

```python
# calculate accuracy of the model
(df.Actual == df.Predicted).sum() / df.shape[0]
```

**Model Accuracy:**  
As you can see, the accuracy for this model is 23.8%, which is slightly lower than the baseline. I will now try to get a sense of how far off the predictions are from the actual retweets.

```python
# find average absolute difference between actual and predicted
(abs(df.Actual-df.Predicted)).mean()
```

```python
# find median absolute difference between actual and predicted
(abs(df.Actual-df.Predicted)).median()
```

**Note:**  
With at least half of the predictions being within 1 retweet of the actual number, I'd like to see the percentage breakdown for other differences. I will create a new dataframe consisting of an error range (i.e. within 1 retweet of actual, within 2 retweets of actual, etc.) and the percentage of predictions that fall within that range.

```python
# create a list of percentages rounded to 1 decimal place
scores = []
for idx in range(6):
    score = (abs(df.Actual - df.Predicted) <= idx).sum() / df.shape[0]
    scores.append(float(format(score*100, '.1f')))

# create the dictionary and convert to a dataframe
df_accuracy = pd.DataFrame({
    'error_range': range(6),
    'percent': scores
})
df_accuracy
```

```python
from matplotlib import rcParams
fig, ax = plt.subplots(figsize=(10,6))

# line plot
ax.plot(df_accuracy.error_range, df_accuracy.percent, lw=2)

# scatter plot to clearly identify the points
ax.scatter(df_accuracy.error_range, df_accuracy.percent, s=75)

# add title and add space between it and graph
ax.set_title('Percent of Predictions Within Given Number of Actual', fontsize=20)
rcParams['axes.titlepad'] = 15

# label the axis and add space between them and graph
ax.set_xlabel('Difference in Actual and Predicted', fontsize=16)
ax.xaxis.labelpad = 15
ax.set_ylabel('Percent of Predictions', fontsize=16)
ax.yaxis.labelpad = 15

# change labels of x-axis tick marks
plt.xticks(range(6), ['Exact', 'Within 1', 'Within 2', 'Within 3',
                      'Within 4', 'Within 5'], fontsize=12)
plt.yticks(fontsize=12)

# label the points with the percentages
[ax.annotate('{}%'.format(xy[1]), xy=(xy[0],xy[1]-4), size=16)\
 for xy in zip(df_accuracy.error_range, df_accuracy.percent)]
plt.show()
```
